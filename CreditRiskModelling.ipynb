```python
#Define the woe_discrete functions
def woe_discrete(df,discrete_variable_name,dependant_variable):
    df=pd.concat([df[discrete_variable_name],df[dependant_variable]],axis=1)
    df.groupby(df.columns.values[0],as_index=False).count()
    df.groupby(df.columns.values[0],as_index=False).mean()
    df=pd.concat([df.groupby(df.columns.values[0],as_index=False).count(),
              df.groupby(df.columns.values[0],as_index=False).mean()],axis=1)
    df=df.iloc[:,[0,1,3]]
    df.columns=[df1.columns.values[0],'n-obs','prop_good']
    df['prop_n_obs']=df['n-obs']/df['n-obs'].sum()
    df['n_good'] = df['n-obs']*df['prop_good']
    df['n_bad']=df['n-obs']*(1-df['prop_good'])
    df['prop_n_good']=df['n_good']/df['n_good'].sum()
    df['prop_n_bad']=df['n_bad']/df['n_bad'].sum()
    df['WoE']=np.log(df['prop_n_good']/df['prop_n_bad'])
    df=df.sort_values(['WoE'])
    df=df.reset_index(drop=True)
    df['IV']=(df['prop_n_good']-df['prop_n_bad'])*df['WoE']
    df['IV']=df['IV'].sum()
    return(df)
```

```python
#Function to define a plot
def plot_by_woe(df_woe,rotation=0):
    #convert x to array
    #The apply() function is used to apply a function along an axis of the DataFrame.
    x=np.array(df_woe.iloc[:,0].apply(str))
    y=df_woe['WoE']
    plt.figure(figsize=(18,6))
    plt.plot(x,y,marker='o',linestyle='--',color='k')
    plt.xlabel(df_woe.columns[0])
    plt.ylabel("Weight of evidence by " + df_woe.columns[0])
    plt.xticks(rotation=rotation)
    
```
